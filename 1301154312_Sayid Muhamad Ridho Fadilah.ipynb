```python
#import Library
import os
import json
import nltk
from nltk import word_tokenize
from nltk.util import ngrams
```

```python
#melakukan import dataset dari komputer dengan file ekstensi json
path_to_json = 'dataset/'
json_files = [pos_json for pos_json in os.listdir(path_to_json) if pos_json.endswith('.json')]
isi = []
for i in json_files:
    with open('dataset\\'+i , encoding='utf-8') as json_data:
        d = json.load(json_data)
        isi.append("o "+ d['content']+ " oo")
```

```python
#tokenisasi
tokens =[]
for i in isi:
    tokens.append(nltk.word_tokenize(i.lower()))

#menghilangkan stopword    
stopword = [".",",","(",")",":",";","?","/","-","_","--","``","\'","\""]
token = []
for i in tokens:
    temp = []
    for j in i:
        if j not in stopword:
            temp.append(j)
    token.append(temp)
        
#menghitung jumlah setiap kata
korpus = {}
totalkata = 0
for i in token:
    for j in i:
        if j in korpus:
            totalkata += 1
            korpus[j] += 1
        else:
            korpus[j] = 1
print (korpus)
```

```python
#melakukan tokenisasi bigram
tokenBigram = []
for i in token:
    tokenBigram.append(list(ngrams(i,2)))
```

```python
#membuat bigram model dengan menghitung jumlah token bigram
bigramModel = {}
allBigram = []
for i in tokenBigram:
    for j in i:
        if j in bigramModel:
            bigramModel[j] += 1
        else:
            allBigram.append(j)
            bigramModel[j] = 1
bigramModel    
```

```python
#menghitung nilai frekuensi kemunculan bigram
freqBigram = {}
for i in bigramModel:
    freqBigram[i] = bigramModel[i]/korpus[i[0]]

freqBigram
```

```python
#menghitung perplexity dari sebuah kalimat
def CalculatePerplexity(kalimat, newFreq):
    nilai = 1
    count = 0
    for i in kalimat:
        nilai *= 1/(newFreq[i])
        count += 1
    return nilai**(1/count)
```
